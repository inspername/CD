```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
dados = pd.read_csv('PES2014.csv')
```

```python
tamanho = dados.shape[0]
data =  pd.DataFrame()
```

```python
#aposentados = dados[(dados.V9122 == 2)]
#internet = dados.V06111.astype('category')
aposentados = dados.V9122.astype('category')
sexo = dados.V0302.astype('category')
lerEscrever = dados.V0601.astype('category')
idade = dados.V8005.astype('float')
curso = dados.V6007.astype('category')

```

```python
#Sabe ler e escrever

graphLerEscrever = pd.crosstab(lerEscrever,sexo[aposentados==2])

graphLerEscrever = graphLerEscrever.iloc[:2,0:2]

graphLerEscrever.columns = ['masculino','feminino']

graphLerEscrever.index = ['sim','não']

plot = graphLerEscrever.plot(kind='pie', title='Alfabetização',autopct='%.1f',figsize=(11,5),subplots=True)
```

Análisando do nível de ensino dos aposentados, observa-se que percentualmente, a quantidade de homens alfabetizados é ligeiramente maior que a de mulheres, e, consequentemente, a quantidade de homens analfabetos é ligeiramente menor que a de mulheres.

```python
#Nível de escolaridade Masculino:

graphCurso = pd.crosstab(curso,sexo[aposentados == 2],margins=True)




graphCurso.columns = ['masculino','feminino','All']

graphCurso.index = ['Elementar (primário)','Médio 1º ciclo (ginasial, etc.)',
'Médio 2º ciclo (científico, clássico, etc.)','Regular do ensino fundamental ou do 1º grau',
'Regular do ensino médio ou do 2º grau',
'Educação de jovens e adultos ou supletivo do ensino fundamental ou do 1º grau',
'Educação de jovens e adultos ou supletivo de ensino médio ou do 2º grau',
'Superior de graduação','Mestrado ou doutorado','Alfabetização de jovens e adultos','Creche',
'Classe de alfabetização - CA','Maternal, jardim de infância etc','All']

graphCursoM = graphCurso.iloc[:13,0]

plot = graphCursoM.plot(kind='pie',title='Nível de Escolaridade Masculino',autopct='%.1f',figsize=(21, 20))
```

```python
#Nível de Escolaridade Feminino:

graphCursoM = graphCurso.iloc[:13,1]

plot = graphCursoM.plot(kind='pie',title='Nível de Escolaridade Feminina',autopct='%.1f',figsize=(21, 20)) 
```

Com base na análise da formação acadêmica dos aposentados participantes da pesquisa, é visível que, embora a divisão da população por nível de escolaridade seja relativamente semelhante entre homens e mulheres, o percentual de mulheres formadas no ensino superior, quando comparado ao de homens, encontra-se maior.

```python
#Idade vs. Sexo:

faixaEtária =  np.arange(10,121,5)

idadeAposentados= pd.cut(idade[aposentados == 2], faixaEtária)

variavel = idadeAposentados.value_counts((True)*100)
variavel = variavel.round(decimals=0)

plot = idade[sexo == 2][aposentados == 2].plot.hist(bins=faixaEtária,color = "indigo",title = 'Homens por Faixa Etária')
```

```python
plot = idade[sexo == 4][aposentados == 2].plot.hist(bins=faixaEtária,color = "pink",title='Mulheres por Faixa Etária')
```

Como pode ser observado com base nos gráficos gerados a respeito das populações de homens e mulheres aposentados que fizeram parte da pesquisa, conclui-se que o número de indivíduos relacionados às suas idades, apresentam-se semelhantes para ambos os sexos.
